```python
# Неделя 1: проверка интерполятора вязкости
import sys
sys.path.append('.')
import matplotlib.pyplot as plt
import numpy as np
from src.fluid import Fluid

# Данные студента Али Иван (введены вручную из hw2_data.csv)
xa = 0.3745      # мольная доля N2
xy = 0.9507      # мольная доля CO2
rho_c = 0.6799   # критическая плотность, кг/м³
M_CH4 = 0.01604
M_N2 = 0.02801
M_CO2 = 0.04401
yCH4 = 1 - xa - xy
M = yCH4 * M_CH4 + xa * M_N2 + xy * M_CO2
T_res = 310.0    # пластовая температура, К

# Создание объекта флюида
fluid = Fluid(M, rho_c, xa, xy, T_res)

# Расчёт вязкости в диапазоне давлений 1..200 атм
P_range = np.linspace(1, 200, 200)
mu_vals = [fluid.mu(p) for p in P_range]

# Построение графика
plt.figure(figsize=(8,5))
plt.plot(P_range, mu_vals, 'b-', linewidth=2)
plt.xlabel('Давление, атм')
plt.ylabel('Вязкость, сП')
plt.title('Зависимость вязкости газа от давления')
plt.grid(True)
plt.show()
```
